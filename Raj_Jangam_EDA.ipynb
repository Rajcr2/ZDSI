```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
customer_df = pd.read_csv('Customers.csv')
product_df = pd.read_csv('Products.csv')
transaction_df = pd.read_csv('Transactions.csv')
```

```python
customer_df.head()
```

```python
product_df.head()
```

```python
transaction_df.head()
```

```python
ct_df = pd.merge(customer_df, transaction_df, on=['CustomerID'])
ct_df
```

```python
pt_df = pd.merge(product_df, transaction_df, on=['ProductID'])
pt_df
```

```python
df = pd.merge(pt_df, customer_df, on=['CustomerID'])
df
```

```python
df.drop(columns=['Price_y'],inplace = True)
df
```

```python
# 2. Derive at least 5 business insights from the EDA.
```

```python
# Business Insight - 1
# 1. Top Performing Region
```

```python
tpr = df
region_quantity = tpr.groupby('Region')['Quantity'].sum()

# Plotting
region_quantity.plot(kind='bar', color='skyblue', figsize=(6, 4))
plt.xlabel('Region')
plt.ylabel('Total Quantity')
plt.title('Top Performing REGION')
plt.show()
```

```python
tpr = df
region_quantity = tpr.groupby('Region')['Quantity'].sum()

# Plotting
region_quantity.plot(kind='barh', color='skyblue', figsize=(6, 4))
plt.xlabel('Total Quantity')
plt.ylabel('Region')
plt.title('Top Performing REGION')
plt.show()
```

```python
# Business Insight - 2
# 2. Most Productive Region
```

```python
mpr = df
region_sales = df.groupby('Region')['TotalValue'].sum()

region_sales.plot(kind='barh', color='orange', figsize=(6, 4))
plt.xlabel('Total Value (USD)')
plt.ylabel('Region')
plt.title('Most Productive REGION')
plt.show()
```

```python
# Business Insight - 3
# 3. Most Required Product_Category (Region-Wise)
```

```python
mrpr = df

category = {'Books': 1, 'Electronics': 2, 'Clothing': 3, 'Home Decor': 4}
mrpr['ctg'] = mrpr['Category'].map(category)
mrpr
```

```python
mrpr = df
product_in_demand = mrpr.groupby(['Region','ctg'])['Quantity'].sum()

product_in_demand.plot(kind='barh', color='skyblue', figsize=(6, 4))
plt.xlabel('Products (Quantity)')
plt.ylabel('Category (Region)')
plt.title('Most Required Product_Category (Region-Wise)')
plt.show()
```
